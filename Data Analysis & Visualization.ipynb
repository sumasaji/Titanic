# Outlier

An outlier is a data point in a data set that is distant from all other observations. A data point that lies outside the overall observation of the dataset.

**What are the impacts of having outliers in a dataset?**
1) It causes various problems during our statistical analysis like accuracy will compromize leading to less accuarate prediction

2) It may cause a significant impact on the mean and standard deviation
